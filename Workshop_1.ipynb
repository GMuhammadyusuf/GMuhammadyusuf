## __Q3.__ What is the average contribution to capital-gain of each sex and occupation category? ##

```python
# Write your code here.
import pandas as pd
data = pd.read_csv('adult.csv')
data = data.sample(n=30000, random_state=448)
data = data.drop(['numbers'], axis=1)
avg_capital_gain = data.groupby(['sex', 'occupation'])['capital-gain'].mean()
print(avg_capital_gain)
```

Using Python/Pandas: Group data by sex & occupation, then average the capital-gain column. Dataset: 30K sampled rows from U.S. census-like data
